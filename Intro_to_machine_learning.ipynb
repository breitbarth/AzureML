## Training a Classification Model

## K-Nearest Neighbors

k-NN finds the k number of nearest data points and makes a educated guess based on the classifications of the nearest datapoint.

![](https://github.com/sagecodes/intro-machine-learning/raw/master/irisknn.png)

```
# Import neighbors model
from sklearn import neighbors

# Assign Data and labels to x & y values
x = iris.data
y = iris.target

# Assign model to variable
# Assign k values to model
knn = neighbors.KNeighborsClassifier(n_neighbors=3)
```

```
# Train the model by passing in the data and features
knn.fit(x, y)
```

```

# 0 = setosa' 1 = 'versicolor' 2 = 'virginica']
flower_types = ['setosa','versicolor','virginica']

# What kind of iris has 3cm x 5cm sepal and 4cm x 2cm petal?

# Data: sepal length, sepal width, petal length, petal width

# change flower values below to see the prediction. 
test_flower = [6.7, -99, 999, 2342]

result = knn.predict([test_flower])

# print index of prediction
print(result)

# Print predicted flower type
print(flower_types[int(result)])

# Print the probability
knn.predict_proba([test_flower])
```

```
# Make the output easier to read.

pred = knn.predict_proba([test_flower])
preddf = pd.DataFrame(pred, columns=iris.target_names)
preddf
```

```
# Lets make it a chart so we can read even more easily!
preddf.plot(kind='bar');
```

There is a problem now... we don't know how well our model actually trained!

#### Training & Testing the Model (The right way)

- Train Set
- Validation Set
- Test Set


```
# import train test split from sklearn
from sklearn.model_selection import train_test_split

# create the test / training data
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.2,
                                                    random_state=21, 
                                                    stratify=y)


knn = neighbors.KNeighborsClassifier(n_neighbors=3)

# Train model on training data set
knn.fit(x_train, y_train)

# Predict the labels on test data sset
y_pred = knn.predict(x_test)
```

```
# Print the list of predictions
print("Test Predictions: \n {}".format(y_pred))
print("True Values: \n {}".format(y_test))
```

```
# Print accuracy
knn.score(x_test, y_test)
```

Why is score (Accuracy) not always such a good measurment

Confusion Matrix

```
# confusion Matrix
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, y_pred)

# L
# a
# b
# e
# l
##### Predictced

cm
```

```
sns.heatmap(cm, annot = True)
```

Classification report

```
print(classification_report(y_test, y_pred))
```

### Train Test Split

Lets do this BEFORE we train this time!

```
# import Train Test Split from sklearn
from sklearn.model_selection import train_test_split
```

```
#create train & test datasets
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.2,
                                                    random_state=4)
```

```
# Assign LinearRegression model to a variable
reg = linear_model.LinearRegression()
```

```
# train the LinearRegression model with training data
reg.fit(x_train, y_train)

# Make predictions on the test data using trained model
y_pred = reg.predict(x_test)
```

```
#print out the predictions
print("Test Predictions: \n {}".format(y_pred))
```

```
# Print accuracy
reg.score(x_test, y_test)
```

```
#plot out predicted and actual prices

import matplotlib.pyplot as plt

predicted = reg.predict(x_test)
expected = y_test

plt.figure(figsize=(8, 5))
plt.scatter(expected, predicted)
plt.xlabel('Actual Price ($1000s)')
plt.ylabel('Predicted price ($1000s)')
```
